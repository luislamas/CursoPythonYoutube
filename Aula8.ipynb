Inicialmente vamos importar as bibliotecas relevantes

```python
import numpy as np
from matplotlib import pyplot as plt
```

Existem basicamente dois tipos de gráficos tridimensionais: Os gráficos de linhas (ou de funções) e os gráficos de malhas (ou meshs). Vamos iniciar nosso estudo com gráficos de funções.

```python
plt.figure()
plt.axes(projection='3d')
```

```python
zdata = np.random.random(100)
xdata = np.random.random(100)
ydata = np.random.random(100)


ax = plt.axes(projection='3d')
ax.scatter3D(xdata, ydata, zdata, c=zdata, cmap='Greens')

ax.azim=-10
ax.dist = 10
ax.elev = 0
```

```python
zline = np.linspace(0, 15, 1000)
xline = np.sin(zline)
yline = np.cos(zline)

ax = plt.axes(projection='3d')
ax.plot3D(xline, yline, zline)
```

Vamos agora abordar gráficos de malhas (mesh). Para esse tipo de gráficos, é necessário inicialmente criar uma malha de duas dimensões. Uma forma de se fazer isso é utilizando-se o comando "np.meshgrid"

```python
a, b, c, d, e, f, g, h = [-327.513, -14.805, -468.928, 0.164638, 525.412, -604.173, 595.482, 0.156]

x, y = np.meshgrid(np.linspace(-2, 2, 105), np.linspace(-2,2, 105))
ZZ = a*(b-x)**2 * np.exp(-x**2-(y+c)**2) - d*(x/e - x**3 - y**5)* np.exp(-x**2-y**2)-f*np.exp(-(x+g)**2-y**2) + h

ax = plt.axes(projection='3d')
ax.plot_surface(x,y,ZZ)
```

```python
ax = plt.axes(projection='3d')
ax.plot_surface(x,y,ZZ, cmap = 'viridis')
#plt.colorbar(ax)
ax.set_xlabel('$X$ axis', fontsize=20)
ax.set_ylabel('$Y$ axis', fontsize=20)
ax.set_zlabel('Porosity', fontsize=20)
ax.xaxis.labelpad=20
ax.yaxis.labelpad=20
ax.zaxis.labelpad=20
```

Detalhes sobre colormaps podem ser encontrados em https://matplotlib.org/stable/tutorials/colors/colormaps.html

```python
fig = plt.figure(figsize=(10,5))
ax = plt.axes(projection='3d')
plot = ax.plot_surface(x,y,ZZ, cmap = 'viridis')
#plt.colorbar(ax)
#ax.set_xlabel('$X$ axis')
#ax.set_ylabel('$Y$ axis')
#ax.set_zlabel('Porosity')

#ax.xaxis.labelpad=20
#ax.yaxis.labelpad=20
#ax.zaxis.labelpad=20
fig.colorbar(plot, label='Porosity')
```

```python
fig = plt.figure(figsize=(10,5))
plt.imshow(ZZ)
fig.colorbar(plot, label='Porosity')
```

```python
a, b, c, d, e, f, g, h = [-327.513, -14.805, -468.928, 0.164638, 525.412, -604.173, 595.482, 0.156]

x, y = np.meshgrid(np.linspace(-2, 2, 105), np.linspace(-2,2, 105))
ZZ = a*(b-x)**2 * np.exp(-x**2-(y+c)**2) - d*(x/e - x**3 - y**5)* np.exp(-x**2-y**2)-f*np.exp(-(x+g)**2-y**2) + h

ZZ = ZZ[:-1, :-1]

fig = plt.figure(figsize=(10,5))
ax = plt.axes()

c = ax.pcolormesh(x, y, ZZ, cmap='RdBu', vmin=ZZ.min(), vmax=ZZ.max())
ax.axis([x.min(), x.max(), y.min(), y.max()])
cbar = fig.colorbar(c, ax=ax)
cbar.set_label('Porosity', size=20)
cbar.ax.tick_params(labelsize=20)
ax.set_xlabel('Eixo X', fontsize=20)
ax.set_ylabel('Eixo Y')
ax.tick_params(labelsize=15)
#fig.show()
```
