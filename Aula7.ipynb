Nessa aula vamos entender como é possível construir e ilustrar gráficos utilizando a biblioteca matplotlib do python. Com essa biblioteca vamos fazer gráficos 2D, além de usar recursos simples de formatação do gráfico.

https://matplotlib.org/

Inicialmente devemos importar as bibliotecas necessárias

```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
y = [0, 0.48, 0.84, 1, 0.91, 0.6, 0.14]
plt.plot(y)
```

Para mudar o tipo da linha e do ponto, pode-se usar alguma combinação de valores de cor, marcadores e tipo de linha.

https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html#matplotlib.pyplot.plot

```python
x = np.linspace(0, 10, 100)
y = np.cos(x)
z = np.sin(x)
```

```python
plt.plot(x,y, x,z)
plt.xlabel('Angulo (rad)')
plt.ylabel('sin / cos')
plt.title('Funções periódicas')
plt.show()
```

```python
plt.plot(x,y,'8k-', label = 'seno')
plt.plot(x,z,'Dm:', label = 'cosseno')
plt.xlabel('Angulo (rad)')
plt.ylabel('sin ou cos')
plt.title('Funções periódicas')
plt.legend()
plt.show()
```

### Scatter plot

```python
x = np.random.rand(100)
y = np.random.rand(100)

plt.scatter(x, y)
plt.title('Exemplo de Scatter Plot')
plt.show()
```

```python
plt.plot(x, y, 'bo')
```

### Gráfico de pizza

```python
valores = [6, 4, 8]
nomes = ['X', 'Y', 'Z']
cores = ['r', 'm', 'y']
plt.pie(valores, labels = nomes, colors = cores, startangle=90)
plt.title('Exemplo de grafico de Pizza')
plt.show()
```

### Histograma

```python
sequencia_aleatoria = np.random.randn(10000,1)
plt.hist(sequencia_aleatoria, bins=30)
plt.title('Exemplo de Histograma')
plt.show()
```

### Barra de erros

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
#dy = 0.1
dy = 0.1*y

plt.errorbar(x, y, dy, errorevery=5, ecolor = 'k')
plt.show()
```

### Gráfico de barras

```python
y = [3, 10, 7, 5, 3, 4, 6, 8]
x = range(len(y))
plt.bar(x, y, color='r')
plt.title('Exemplo de gráfico de barras')
plt.show()
```

### Linhas verticais e horizontais

```python
x = np.linspace(0, 50, 200)
y = np.sin(x) * np.exp(-x/10)
plt.plot(x,y, 'k')
plt.hlines(0.5, xmin=-10, xmax=70)
plt.vlines(30, ymin=-2, ymax=2, color='k', linestyles =':')
plt.axis([0, 50, -1, 1])
```

### Gráfico Polar

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.polar(x,y,'k')
plt.show()
```

### Escalas Logarítmicas

```python
x = np.arange(0.01, 20.0, 0.01)
plt.plot(x, np.exp(-x))
plt.semilogy(x, np.exp(-x))
plt.grid(True, which='both')
plt.minorticks_on()
```

```python
x = np.arange(0.01, 20.0, 0.01)
plt.figure()
#plt.plot(x, np.sin(x))
plt.semilogx(x, np.sin(x))
```

```python
x = np.logspace(-1,2)
y1 = 10**x
y2 = 1/10**x
plt.loglog(x,y1,x,y2)
plt.title('Exemplo de loglog')
plt.show()
```

```python
x = np.linspace(0, 10, 100)

fig, axs = plt.subplots(2, 2, sharex=True)
axs[0, 0].plot(x, np.cos(x))
axs[0, 0].set_title('cosseno')
axs[0, 1].plot(x, np.exp(-x))
axs[0, 1].set_title('Exponencial')
axs[1, 0].plot(x, x**2)
axs[1, 0].set_title('x**2')
axs[1, 1].plot(x, 1/x)
axs[1, 1].set_title('1/x')

```
